```python
import pandas as pd
```

```python
from matplotlib import pyplot as pt
```

```python
sample_data = pd.read_csv('bank1.csv')
```

```python
sample_data
```

```python
pt.bar(sample_data['age'],sample_data['balance'])
pt.xlabel('AGE GROUP')
pt.ylabel('ACCOUNT BALANCE')
pt.title("BAR GRAPH")
pt.xticks(rotation = 45)
pt.show()
```

```python

pt.bar(sample_data['job'],sample_data['balance'])
pt.xlabel('JOB-TYPE')
pt.ylabel('ACCOUNT BALANCE')
pt.title("BAR GRAPH2")
pt.xticks(rotation = 45)
pt.show()
```

```python
pt.figure(figsize=(6,6))
pt.pie(sample_data['age'],labels=sample_data['balance'],autopct='%1.1f%%',startangle=140)
pt.title('Pie chart')
pt.show()
```

```python
age = sample_data['age']
balance = sample_data['balance']
pt.plot(age,balance)
pt.xlabel('Age-GROUP')
pt.ylabel('ACCOUNT BALANCE')
pt.title('LINE CHART AGE/BALANCE')
pt.show()
          
```
